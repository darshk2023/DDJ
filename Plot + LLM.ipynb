```python
from langchain.text_splitter import CharacterTextSplitter

text_splitter = CharacterTextSplitter.from_tiktoken_encoder(
    chunk_size=7500, chunk_overlap=100
)
doc_splits = text_splitter.split_documents(docs_list)
```

```python
import tiktoken

encoding = tiktoken.get_encoding("cl100k_base")
encoding = tiktoken.encoding_for_model("gpt-3.5-turbo")
for d in doc_splits:
    print("The document is %s tokens" % len(encoding.encode(d.page_content)))
```

```python
import os

from langchain_community.vectorstores import Chroma
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableLambda, RunnablePassthrough
from langchain_nomic import NomicEmbeddings
from langchain_nomic.embeddings import NomicEmbeddings

# Add to vectorDB
#chroma = database on my laptop, look up how to put my own documents in chroma
#manually find the scouting reports
#lang chain pdf reader, get the text out of it and put the text in chroma

vectorstore = Chroma.from_documents(
    documents=doc_splits,
    collection_name="rag-chroma",
    embedding=NomicEmbeddings(model="nomic-embed-text-v1"),
)
retriever = vectorstore.as_retriever()
```

```python
#how to do self reflection in langchain
from langchain_community.chat_models import ChatOllama
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

# Prompt
template = """Answer the question based only on the following context:
{context}

Question: {question}
"""
prompt = ChatPromptTemplate.from_template(template)

# LLM API
#model = ChatOpenAI(temperature=0, model="gpt-4-1106-preview")

# Local LLM
ollama_llm = "mistral:instruct"
model_local = ChatOllama(model=ollama_llm)

# Chain
chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt
    | model_local
    | StrOutputParser()
)
```

```python
#!pip install customtkinter
import customtkinter as ctk
import tkinter as tk
from tkinter import messagebox
import webbrowser
import tempfile
import os
import plotly.io as pio
import pandas as pd
import plotly.graph_objects as go

# Assuming your data is loaded here
player_data = pd.read_csv('NBA_Players.csv')
team_data = pd.read_csv('TeamPerGame.csv')

# Placeholder for your plotting functions
def create_spider_plot_percentiles_plotly_player(player_data, player1_name, player2_name, percentile_columns, years_input):
    """
    Create an interactive spider plot for specified seasons of the specified players, with each player's performance in different colors.
    The axes are based on percentile ranks relative to other NBA players using Plotly. Percentile ranks are calculated for specified columns.

    Parameters:
    - player_data (DataFrame): The data containing NBA player stats.
    - player1_name, player2_name (str): The names of the players for whom the spider plot is to be created.
    - percentile_columns (list): List of metrics for which percentile ranks are calculated and included in the spider plot.
    - years_input (str): A single year, a range of years (e.g., "2020-2023"), or a comma-separated list of years (e.g., "2020,2021,2023") for which the comparison is to be made.

    Returns:
    - Plotly Figure: This function will return a Plotly figure which can be displayed using fig.show().
    """
    # Calculate percentiles for the specific metrics within the dataset for all teams relative to the year
    for col in percentile_columns:
        # Group by 'year' and calculate percentiles within each group
        player_data[col + '_percentile'] = player_data.groupby('season')[col].rank(pct=True)

    def filter_data_for_player(player_name):
        if years_input == 'all':
            return player_data[player_data['player'] == player_name]
        elif '-' in years_input:
            start_year, end_year = map(int, years_input.split('-'))
            years = list(range(start_year, end_year + 1))
            return player_data[(player_data['player'] == player_name) & (player_data['season'].isin(years))]
        else:
            years = [int(year) for year in years_input.split(',')]
            return player_data[(player_data['player'] == player_name) & (player_data['season'].isin(years))]

    player1_data = filter_data_for_player(player1_name)
    player2_data = filter_data_for_player(player2_name)

    if player1_data.empty or player2_data.empty:
        if player1_data.empty:
            print(f"No data found for player: {player1_name} in the specified years: {years_input}")
        if player2_data.empty:
            print(f"No data found for player: {player2_name} in the specified years: {years_input}")
        return None

    fig = go.Figure()

    # Adding radar charts for each player and season
    for player_data, player_name, color in zip([player1_data, player2_data], [player1_name, player2_name], ['blue', 'red']):
        if not player_data.empty:
            for index, season_data in player_data.iterrows():
                fig.add_trace(go.Scatterpolar(
                    r=season_data[[col + '_percentile' for col in percentile_columns]].tolist(),
                    theta=[col for col in percentile_columns],
                    fill='toself',
                    name=f"{player_name}: {season_data['season']}",
                    line=dict(color=color)
                ))

    fig.update_layout(
        polar=dict(
            radialaxis=dict(
                visible=True,
                range=[0.0, 1.0],
                dtick=0.10
            )),
        showlegend=True,
        title=f'{player1_name} vs {player2_name} Performance Comparison'
    )

    return fig
    pass

def create_spider_plot_percentiles_plotly_team(team_data, team1_name, team2_name, percentile_columns, years_input):
    """
    Create an interactive spider plot for specified seasons of the specified teams, with each team's performance in different colors.
    The axes are based on percentile ranks relative to other NBA teams using Plotly. Percentile ranks are calculated for specified columns.

    Parameters:
    - team_data (DataFrame): The data containing NBA team stats.
    - team1_name, team2_name (str): The names of the teams for whom the spider plot is to be created.
    - percentile_columns (list): List of metrics for which percentile ranks are calculated and included in the spider plot.
    - years_input (str): A single year, a range of years (e.g., "2020-2023"), or a comma-separated list of years (e.g., "2020,2021,2023") for which the comparison is to be made.

    Returns:
    - Plotly Figure: This function will return a Plotly figure which can be displayed using fig.show().
    """
    # Calculate percentiles for the specific metrics within the dataset for all teams relative to the year
    for col in percentile_columns:
        # Group by 'season' and calculate percentiles within each group
        team_data[col + '_percentile'] = team_data.groupby('season')[col].rank(pct=True)

    def filter_data_for_team(team_name):
        if years_input == 'all':
            return team_data[team_data['team'] == team_name]
        elif '-' in years_input:
            start_year, end_year = map(int, years_input.split('-'))
            years = list(range(start_year, end_year + 1))
            return team_data[(team_data['team'] == team_name) & (team_data['season'].isin(years))]
        else:
            years = [int(year) for year in years_input.split(',')]
            return team_data[(team_data['team'] == team_name) & (team_data['season'].isin(years))]

    team1_data = filter_data_for_team(team1_name)
    team2_data = filter_data_for_team(team2_name)

    if team1_data.empty or team2_data.empty:
        if team1_data.empty:
            print(f"No data found for team: {team1_name} in the specified years: {years_input}")
        if team2_data.empty:
            print(f"No data found for team: {team2_name} in the specified years: {years_input}")
        return None

    fig = go.Figure()

    for team_data, team_name, color in zip([team1_data, team2_data], [team1_name, team2_name], ['blue', 'red']):
        if not team_data.empty:
            for index, season_data in team_data.iterrows():
                fig.add_trace(go.Scatterpolar(
                    r=season_data[[col + '_percentile' for col in percentile_columns]].tolist(),
                    theta=[col for col in percentile_columns],
                    fill='toself',
                    name=f"{team_name}: {season_data['season']}",
                    line=dict(color=color)
                ))

    fig.update_layout(
        polar=dict(
            radialaxis=dict(
                visible=True,
                range=[0, 1],
                dtick=0.10  # Set tick marks at increments of 0.10
            )),
        showlegend=True,
        title=f'{team1_name} vs {team2_name} Performance Comparison'
    )

    return fig


# AnalysisApp definition
class AnalysisApp(ctk.CTk):
    def __init__(self):
        super().__init__()
        self.title('Player vs. Team Analysis')
        self.geometry('500x350')

        self.choice_frame = ctk.CTkFrame(self)
        self.choice_frame.pack(pady=20, padx=20, fill='both', expand=True)

        self.player_button = ctk.CTkButton(self.choice_frame, text="Player Analysis", command=lambda: self.show_form('player'))
        self.player_button.pack(pady=10, fill='x')

        self.team_button = ctk.CTkButton(self.choice_frame, text="Team Analysis", command=lambda: self.show_form('team'))
        self.team_button.pack(pady=10, fill='x')

        self.form_frame = ctk.CTkFrame(self)
        self.name1_entry = ctk.CTkEntry(self.form_frame)
        self.name2_entry = ctk.CTkEntry(self.form_frame, placeholder_text="Second Name")
        self.years_entry = ctk.CTkEntry(self.form_frame, placeholder_text="Year or Range of Years")
        
        # Button frame for Submit and Back buttons
        self.button_frame = ctk.CTkFrame(self.form_frame)
        self.button_frame.pack(side="bottom", pady=10, fill='x', expand=True)
        
        self.back_button = ctk.CTkButton(self.button_frame, text="Back", command=self.go_back)
        self.back_button.pack(side='left', padx=10, expand=True)

        self.submit_button = ctk.CTkButton(self.button_frame, text="Submit", command=self.submit)
        self.submit_button.pack(side='right', padx=10, expand=True)


    def show_form(self, analysis_type):
        self.analysis_type = analysis_type
        self.choice_frame.pack_forget()

        placeholder_text = "Player 1 Name" if analysis_type == 'player' else "Team 1 Name"
        self.name1_entry.configure(placeholder_text=placeholder_text)
        self.name2_entry.configure(placeholder_text="Player 2 Name" if analysis_type == 'player' else "Team 2 Name")

        self.form_frame.pack(pady=20, padx=20, fill='both', expand=True)
        self.name1_entry.pack(pady=10, fill='x')
        self.name2_entry.pack(pady=10, fill='x')
        self.years_entry.pack(pady=10, fill='x')
        self.submit_button.pack(pady=20)

    def submit(self):
        name1 = self.name1_entry.get()
        name2 = self.name2_entry.get()
        years = self.years_entry.get()
        
        

        player_data = pd.read_csv('NBA_Players.csv')
        team_data = pd.read_csv('TeamPerGame.csv')
        team_percentile_columns = ['pts_per_game','tov_per_game','blk_per_game','stl_per_game','ast_per_game','drb_per_game','orb_per_game','ft_percent','fta_per_game'] 
        player_percentile_columns = ['usg_percent', 'ows', 'dws', 'obpm', 'dbpm', 'vorp', 'per', 'mp', 'ts_percent','tov_percent', 'blk_percent', 'stl_percent'] 
        

        if self.analysis_type == 'player':
            fig = create_spider_plot_percentiles_plotly_player(player_data, name1, name2, player_percentile_columns, years)
        else:
            fig = create_spider_plot_percentiles_plotly_team(team_data, name1, name2, team_percentile_columns, years)

        if fig:
            temp_file = tempfile.NamedTemporaryFile(delete=False, suffix='.html')
            pio.write_html(fig, file=temp_file.name)
            webbrowser.open('file://' + os.path.realpath(temp_file.name))
        else:
            messagebox.showerror("Error", "No data found or error in generating plot.")

        print(chain.invoke(f'It is {years}, use stats from 2024 ONLY. Imagine you are a coach and need to give advice to {name1} and {name2}. Who do you think is more likely to win in a game against each other. What can the two do to improve, and how can they take advantage of the other’s weaknesses.'))

        self.form_frame.pack_forget()
        self.choice_frame.pack(pady=20, padx=20, fill='both', expand=True)

        
    def go_back(self):
        # Clear entries
        self.name1_entry.delete(0, 'end')
        self.name2_entry.delete(0, 'end')
        self.years_entry.delete(0, 'end')

        # Hide the form frame
        self.form_frame.pack_forget()

        # Show the choice frame again
        self.choice_frame.pack(pady=20, padx=20, fill='both', expand=True)

if __name__ == "__main__":
    app = AnalysisApp()
    app.mainloop()
```
